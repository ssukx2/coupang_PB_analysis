```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# CSV 파일에서 데이터 읽어오기

df = pd.read_csv("C:/Users/attac/OneDrive/바탕 화면/review.csv")

# 리뷰 데이터만 추출
reviews = df["리뷰"].tolist()

# 리뷰 텍스트를 하나의 문자열로 합침
text = ' '.join(reviews)

# 워드 클라우드 생성
wordcloud = WordCloud(font_path='font/NanumGothic.ttf',
                      width=800, height=800, background_color='white').generate(text)

# 워드 클라우드 시각화
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
import pandas as pd
from konlpy.tag import Okt
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# 데이터프레임 생성

# 형태소 분석기 초기화
okt = Okt()

# 불용어 목록 정의
stopwords = ["진짜", "정말", "너무", '은', '는','이','가','도', '이렇게', '에','요','을', '를','인데','으로','로','입니다','합니다','다','했는데','하고','이네','점점','에서','의','이런','들','과','그냥','것','걸','때','했네요','더','원래','게','근데','지만','뭐','이랑','추천','좋은','물론','썼는데','내','쓰고','에요','있을','민','나','안']


# 불용어 제거 함수
def remove_stopwords(text):
    tokens = okt.morphs(text)  # 형태소 분석
    tokens = [word for word in tokens if word not in stopwords]  # 불용어 제거
    return ' '.join(tokens)

# 데이터프레임의 모든 텍스트 열에 불용어 제거 적용
df['cleaned_review'] = df['review'].apply(remove_stopwords)

# 모든 리뷰를 하나의 문자열로 결합
text = ' '.join(df['cleaned_review'])

# 워드 클라우드 생성
wordcloud = WordCloud(font_path='font/c:/windows/fonts/malgunsl.ttf', background_color='white').generate(text)

# 워드 클라우드 시각화
plt.figure(figsize=(10, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')

plt.show()
```
