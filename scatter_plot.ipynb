```python
df_name = 'water'
df = water.copy()
x = df.loc[:49, 'rank']
y = df.loc[:49, 'unit_price']
title = {'water':'생수', "bodywash":"바디워시", "fabric_softner":"섬유유연제","wettissue":"물티슈"}.get(df_name)
```

```python
## 생수
df_name = 'water'
df = water.copy()
x = df.loc[:49, 'rank']
y = df.loc[:49, 'unit_price']
title = {'water':'생수', "bodywash":"바디워시", "fabric_softner":"섬유유연제","wettissue":"물티슈"}.get(df_name)

plt.style.use('ggplot')
# plt.figure(figsize=(15, 8))
# colorlist = [ 'viridis', 'plasma', 'inferno', 'magma' ,'jet','cividis','coolwarm']
# for color in colorlist: 
df.loc[:49,:].plot(x='rank',y='unit_price',kind='scatter',s = 70,  c=df.loc[:49, 'PB'].map({0: 'cornflowerblue', 1: 'salmon'}))


# 선형 회귀 계산
m, b = np.polyfit(x, y, 1)

# 추세선 추가
plt.plot(x, m*x + b, color='darkred', label=f'y = {m:.2f}x + {b:.2f}', ls='--' )

plt.xlabel('Rank')
plt.ylabel('Unit Price')
plt.title(f'{title} 단위 가격과 판매순위의 관계')
# plt.text(50,1000, f'y = {m:.2f}x + {b:.2f}', fontsize=12, color='red')
plt.legend(loc='upper left')
plt.savefig(f'{title}_판매량.png')
plt.show()
```
